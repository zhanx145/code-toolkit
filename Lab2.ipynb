## Loading an Image

```python
import skimage
```

```python
from skimage import io
```

```python
f = io.imread('cat.JPG')
```

```python
f.shape
```

```python
type(f)
```

```python
f[4, 9]
```

```python
plt.imshow(f)
```

```python
from skimage.color import rgb2gray
```

```python
grayscale = rgb2gray(f)
```

```python
plt.imshow(grayscale, cmap=plt.cm.gray)
```

```python
grayscale[4, 9]
```

```python
grayscale.shape
```

```python
def fairey(image_file):
    f=io.imread(image_file)
    grayscale=rgb2gray(f)
    rows=grayscale.shape[0]
    cols=grayscale.shape[1]
    jobama=np.zeros((rows,cols,3))
    for i in range (rows):
        for j in range (cols):
            if grayscale[i,j]>0.75:
                jobama[i,j]=[100,100,255]
            elif grayscale[i,j]>0.5:
                jobama[i,j]=[0,150,180]
            elif grayscale [i,j]>0.25:
                jobama[i,j]=[100,50,0]
            else:
                jobama[i,j]=[0,0,0]
    return plt.imshow(jobama)
            
```

```python
fairey('cat.jpg')
```
